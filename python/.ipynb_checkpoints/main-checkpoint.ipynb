```python
import os
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "../resources/NYC_Dog_Licensing_Dataset.csv"
data = pd.read_csv(csv_path)
data
```

```python
data.groupby(['AnimalGender'], as_index=False).count()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas

# Labels for the sections of our pie chart
labels = ["Female", "Male"]

# The values of each section of the pie chart
sizes = data.groupby(['AnimalGender'], as_index=False).count()['RowNumber']

# The colors of each section of the pie chart
colors = ["thistle", "cornflowerblue"]

# Tells matplotlib to seperate the "Humans" section from the others
explode = (0.1, 0)
plt.title("Distrubution of Female vs Male Dogs")

gender_pie = plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
```
